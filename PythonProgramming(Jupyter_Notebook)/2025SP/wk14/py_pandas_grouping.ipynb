## groupby

- The groupby is one of the most frequently used Pandas functions in data analysis.

- t is used for grouping the data points based on the distinct values in the given column or columns.

- We can then calculate aggregated values for the generated groups. 


```python
cars = ['Toyota','Ford','GMC','Dodge']

df = DataFrame({'brand': np.random.choice(cars,size=20),
                'price':np.random.randint(40000,65000,size=20)})
df
```

```python
df.groupby('brand').count()
```

```python
df.groupby('brand').sum()
```

```python
df.groupby('brand').mean()
```

```python
df.groupby('brand')['price'].sum()
```

```python
df['price'].groupby(df['brand']).sum()
```

```python
rs = df.groupby('brand')['price'].sum()
rs
```

```python
type(rs)
```

```python
rs = df.groupby('brand').sum()
rs
```

```python
type(rs)
```

```python
rs = df.groupby('brand')[['price']].sum()
rs
```

```python
df
```

```python
df.sort_values(by='brand',inplace=True)
df
```

```python
df['cum_sum'] = df.groupby('brand')['price'].cumsum()
df
```
