```python
# two data sets
data = randn(30).cumsum()
data
```

```python
plt.plot(data, 'k--')
plt.plot(data, 'k-', drawstyle='steps-post')
```

```python
#using labels 
plt.plot(data, 'k--', label="Dashed-Line")
plt.plot(data, 'k-', drawstyle='steps-post', label="steps-post")
plt.legend(loc='best') # loc = ["upper left", "lower left", "center right"]
```

```python
# plot multiple columns
d = {'col1': np.random.randint(1,50, size=100), 'col2': np.random.randint(1,50, size=100) * 25}
df = DataFrame(d)
df
     
```

```python
plt.plot(df['col1'],'k-', label='col1')
plt.plot(df['col2'],'ko--', label='col2')
plt.legend(loc='best')
```

```python
#Ticks, Labels, and Legends
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(randn(1000).cumsum())

```

```python
ax.set_xticks(list(range(0,1001,250)))
ax.set_xticklabels(['one','two','three','four','five'], rotation=30, fontsize='small')
ax.set_title('My First Matplotlip - Plot with Ticks')
ax.set_xlabel('Stages')
ax.set_ylabel('Values')
fig
```

```python
props ={'title': 'My First Matplotlip Plot with Ticks',
        'xlabel': 'Stages',
        'ylabel': 'Values'}
ax.set(**props)
fig
```
