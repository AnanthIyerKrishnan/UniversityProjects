```python
df = pd.DataFrame(randn(30).cumsum())
df
```

```python
plt.plot(df, drawstyle='steps-post')
```

```python
plt.plot(df, 'k--')
plt.plot(df,'k-', drawstyle='steps-post')
```

```python
#using labels
plt.plot(df, 'k--', label='Dahsed-Line')
plt.plot(df,'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best') # axes.locs = ["upper left", "lower left", "center right"]
```

```python
#Ticks, Labels, and Legends
fig = plt.figure()
ax = fig.add_subplot(111)
```

```python
ax.plot(randn(1000).cumsum())
```

```python
fig
```

```python
ax.set_xticks([0,250,500,750,1000])
ax.set_xticklabels(['one','two','three','four','five'], rotation=30, fontsize='small')

ax.set_title('My First Matplotlip Plot with Ticks')
ax.set_xlabel('Stages')
ax.set_ylabel('Values')

fig
```

```python
#Use dictionary for title, xlabel, and ylabel
props ={'title': 'My First Matplotlip Plot with Ticks',
        'xlabel':'Stages',
       'ylabel': 'Values'}

ax.set(**props)
fig
```
